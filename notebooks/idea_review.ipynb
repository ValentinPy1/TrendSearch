```python
import pandas as pd

ideas = pd.read_csv("ideas.csv")
users = pd.read_csv("users.csv")
```

```python
ideas_per_user = ideas['user_id'].value_counts().reset_index()
ideas_per_user.columns = ['user_id', 'idea_count']
ideas_per_user = ideas_per_user.merge(users[['id', 'first_name', 'last_name']], left_on='user_id', right_on='id', how='left')
ideas_per_user = ideas_per_user[['id', 'first_name', 'last_name', 'idea_count']]
print(ideas_per_user)
```

```python
# Print original_idea for all users with 10 or less ideas (only rows with non-null original_idea)
ideas_with_counts = ideas['user_id'].value_counts()
eligible_user_ids = ideas_with_counts[ideas_with_counts <= 10].index

for user_id in eligible_user_ids:
    user_ideas = ideas[(ideas['user_id'] == user_id) & ideas['original_idea'].notna()].copy()
    if user_ideas.empty:
        continue
    user_ideas['created_at'] = pd.to_datetime(user_ideas['created_at'].str.replace('"', ''))
    user_ideas = user_ideas.sort_values(by='created_at')
    user_row = users[users['id'] == user_id]
    if not user_row.empty:
        first_name = user_row.iloc[0]['first_name']
        last_name = user_row.iloc[0]['last_name']
        user_display = f"{first_name} {last_name}"
    else:
        user_display = f"(unknown, id={user_id})"
    print(f"User {user_display}:")
    for _, row in user_ideas.iterrows():
        print(f"  {row['created_at']}: {row['original_idea']}")
    print()
```
